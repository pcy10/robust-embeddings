```python
model = BertModel.from_pretrained('bert-base-uncased')
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
import torch
```

```python
def preprocess(text):
    text = text.lower()
    text = re.compile(r'\'<[^>]+>').sub("",text)
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r'https?:\/\/.*[\r\n]*', '', text)
    tokens = text.split(" ")
    tokens_wo_stopwords = [word for word in tokens if not word in stop_words]
    lemmatizer = WordNetLemmatizer()
    lem_words = [lemmatizer.lemmatize(word,pos="n") for word in tokens_wo_stopwords]
    sent = ' '.join(lem_words)
    return sent
```

```python
items = []
grp = []
for i in df.index:
  name =  str(df["Rating"][i]) +  "_"  +  str(df["virality"][i])
  brand = str(df['Review'][i])
  sent = brand
  sent = preprocess(sent.replace("'",''))
  items.append(sent)
  grp.append(name)
```

```python
input_ids = list(map(lambda x: nn.tensor(tokenizer.encode(x))[:512], items))
```

```python
embedd = []
ctr = 0
for i in input_ids:
  if( ctr == 100):
    break;
  ctr+=1
  with torch.no_grad():
    i = i.unsqueeze(0)
    outputs = model(i)
    embedd.append(outputs[0].cpu().detach())
  print(ctr)
```

```python
embedd = np.array(embedd)
```
