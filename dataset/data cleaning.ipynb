```python

from google.colab import drive
drive.mount('/gd')
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.ticker as mtick
import matplotlib.pyplot as plt
df= pd.read_csv('/gd/My Drive/reviews.csv')

```

```python
df = df[['userName' ,'content','score','thumbsUpCount','at']]
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.dropna(inplace = True)
df.isnull().sum()
```

```python
df.drop_duplicates(inplace = True)
df.duplicated().sum()
```

```python
df = df.astype({
    'score':'int16',
    'thumbsUpCount':'int16'
})
df['at'] = pd.to_datetime(df['at'])
df['at'] = df['at'].dt.strftime('%Y')
df.info()
```

```python
df.rename(columns={
    'userName':'User' ,
    'content':'Review',
    'score':'Rating',
    'thumbsUpCount':'Likes',
    'at':'Year'
}, inplace=True)
```

```python
df = df.head(20000)
df.shape
```

```python
for index, row in df.iterrows():
  if row['Likes'] <= 10:
     df.loc[index, 'virality'] = "level 1"
  elif row['Likes'] > 10 and row['Likes'] <= 20:
     df.loc[index, 'virality'] = "level 2"
  elif row['Likes'] > 20 and row['Likes'] <= 30:
     df.loc[index, 'virality'] = "level 3"
  elif row['Likes'] > 30 and row['Likes'] <= 40:
     df.loc[index, 'virality'] = "level 4"
  elif row['Likes'] > 40 and row['Likes'] <= 50:
     df.loc[index, 'virality'] = "level 5"
  elif row['Likes'] > 50 and row['Likes'] <= 6000:
     df.loc[index, 'virality'] = "level 6"
  elif row['Likes'] > 6000 and row['Likes'] <= 7000:
     df.loc[index, 'virality'] = "level 7"
  else:
     df.loc[index, 'virality'] = "level 8"
```
